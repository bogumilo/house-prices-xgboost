# Base Pipeline

### - Preprocessing

```python
from sklearn.impute import KNNImputer, SimpleImputer
from sklearn.preprocessing import OneHotEncoder
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import MinMaxScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer, make_column_selector

# Categorical transformer
cat_transformer = make_pipeline(SimpleImputer(strategy='most_frequent'),
                                OneHotEncoder(sparse = True, handle_unknown='ignore'))

# Numerical transformer
num_transformer = make_pipeline(KNNImputer(n_neighbors=5),
                                MinMaxScaler())
num_col = make_column_selector(dtype_include=['float', 'int'])

# Build pipeline
preproc_baseline = ColumnTransformer([
    ('num_tr', num_transformer, num_col),
    ('cat_tr', cat_transformer, feat_categorical_small)]
)

X = data.drop(['Id', 'SalePrice'], axis=1)
preproc_baseline
```

```python
preproc_baseline_df = pd.DataFrame(preproc_baseline.fit_transform(X), columns = preproc_baseline.get_feature_names_out())
preproc_baseline_df.head(10)
```

```python
shape_preproc_baseline = preproc_baseline_df.shape
shape_preproc_baseline
```

```python
preproc_baseline_df.describe()
```

### - Add estimator

```python
from sklearn.tree import DecisionTreeRegressor

y = data['SalePrice']

pipe_baseline = make_pipeline(preproc_baseline, DecisionTreeRegressor())
pipe_baseline
```

```python
pipe_baseline.get_params()
```

### - Cross-validate

```python
from sklearn.metrics import make_scorer
from sklearn.model_selection import cross_val_score, GridSearchCV

# _score_func(y, y_pred, **kwargs)
def get_rmsle(y, y_pred):
    return np.sqrt(np.mean(np.square(np.log1p(y) - np.log1p(y_pred))))

# minimised
rmsle = make_scorer(get_rmsle, greater_is_better=False)
# maximise
rmsle_neg = make_scorer(get_rmsle, greater_is_better=True)
```

```python
cv_results = cross_val_score(pipe_baseline, X, y, scoring = rmsle_neg, cv=5).mean()
cv_results
```

```python
search_pos = cross_val_score(pipe_baseline, X, y, scoring=rmsle, cv=5)
search_neg = GridSearchCV(
    pipe_baseline,
    param_grid = {
        # Access any component of the Pipeline, as far back as you want
        'decisiontreeregressor__min_samples_leaf': [8, 9, 10, 11, 12],
        'decisiontreeregressor__min_weight_fraction_leaf': [0.0, 0.1],
        'decisiontreeregressor__max_features': [ 1.0, 'sqrt', 'log2'],},
    scoring=rmsle_neg,
    cv=5,
    n_jobs=-1 
    ).fit(X, y)
```

```python
search_pos.mean()
```

```python
search_neg.best_params_
```

```python
score_baseline = search_neg.best_score_
score_baseline
```

### - Predict base

```python
X_test = pd.read_csv('data/test.csv').drop(columns = ['Id'])

y_pred_baseline = search_neg.predict(X_test)
y_pred_baseline
```

# Apply XGBoost

```python
from xgboost import XGBRegressor

xgb = XGBRegressor()

pipe_xgb = make_pipeline(preproc_baseline, xgb)
pipe_xgb
```

```python
xgb_cv = cross_val_score(pipe_xgb,X,y,cv=10, scoring = rmsle_neg, error_score='raise')
xgb_cv
```

```python
xgb_cv.mean()
```

```python
X_test = pd.read_csv('data/test.csv').drop(columns = ['Id'])
y_pred_xgb = pipe_xgb.fit(X, y).predict(X_test)
y_pred_xgb
```

# Evaluate with the 'perfect' solution

```python
solution   = pd.read_csv('data/solution.csv')
y_true     = solution["SalePrice"]

from sklearn.metrics import mean_squared_log_error
RMSLE = np.sqrt( mean_squared_log_error(y_true, y_pred_xgb) )
print("The score is %.6f" % RMSLE )
```
